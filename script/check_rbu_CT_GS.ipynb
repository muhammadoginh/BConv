```python
def DIV2(X,q):
    if (X % 2 == 0):
        X = X >> 1  # Right-shift to divide by 2
    else: # n is odd
        X = (X >> 1) + ((q + 1) >> 1)   # Modular adjustment for odd n
    return X
```

```python
DIV2(140734015810665, 140737518764033)
```

```python
def safe_GS_B0(row):
    return DIV2((int(row['A1']) + int(row['A0'])) % int(row['q']), int(row['q']))

df['Expected_B0'] = df.apply(safe_GS_B0, axis=1)
```

```python
def safe_GS_B1(row):
    return DIV2(((int(row['A0']) - int(row['A1'])) * int(row['C'])) % int(row['q']), int(row['q']))

df['Expected_B1'] = df.apply(safe_GS_B1, axis=1)
```

```python
def safe_CT_B0(row):
    return (int(row['C']) * int(row['A1']) + int(row['A0'])) % int(row['q'])

df['Expected_B0'] = df.apply(safe_CT_B0, axis=1)
```

```python
def safe_CT_B1(row):
    return (int(row['A0']) - int(row['C']) * int(row['A1'])) % int(row['q'])

df['Expected_B1'] = df.apply(safe_CT_B1, axis=1)
```

```python
def safe_mod_mul(row):
    return (int(row['C']) * int(row['A1'])) % int(row['q'])

df['Expected_M'] = df.apply(safe_mod_mul, axis=1)
```
