```python
import pandas as pd
import numpy as ny
```

```python
filename = "/home/moginh/Projects/BConv/BConv.sim/sim_1/behav/xsim/rbu_results.csv"
df = pd.read_csv(filename)
```

```python
# Disable scientific notation for all floats/integers
pd.set_option('display.precision', 0)  # no decimal places
pd.set_option('display.float_format', '{:.0f}'.format)
```

```python
df.head()
```

```python
def safe_MMA(row):
    return ((int(row['A1']) * int(row['C'])) + int(row['A0'])) % int(row['q'])

df['Expected_M'] = df.apply(safe_MMA, axis=1)
df.head(20)
```

```python
def safe_mod_add(row):
    return (int(row['A0']) + int(row['A1'])) % int(row['q'])

df['Expected_BO'] = df.apply(safe_mod_add, axis=1)
```

```python
df.head()
```

```python
def safe_mod_sub(row):
    return (int(row['A0']) - int(row['A1'])) % int(row['q'])

df['Expected_B1'] = df.apply(safe_mod_sub, axis=1)
df.head(6)
```

```python
def safe_MSM(row):
    return (int(row['A0']) - int(row['A1'])) * int(row['C']) % int(row['q'])

df['Expected_M'] = df.apply(safe_MSM, axis=1)
df.head(20)
```

```python
df['Expected_delayed'] = df['Expected_M'].shift(13)
df
```
