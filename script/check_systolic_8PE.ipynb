```python
def systolic(row):
    return ( (int(row['x0']) * int(row['qHat0'])% int(row['q']) 
              + (int(row['x1']) * int(row['qHat1'])) % int(row['q'])
              + (int(row['x2']) * int(row['qHat2'])) % int(row['q'])
              + (int(row['x3']) * int(row['qHat3'])) % int(row['q'])
              + (int(row['x4']) * int(row['qHat4'])) % int(row['q'])
              + (int(row['x5']) * int(row['qHat5'])) % int(row['q'])
              + (int(row['x6']) * int(row['qHat6'])) % int(row['q'])
              + (int(row['x7']) * int(row['qHat7'])) % int(row['q'])) % int(row['q']))

df['Expected_B0'] = df.apply(systolic, axis=1)
```
